Defino las reacciones a simular:
    
$k_1 \hspace{1cm} 2 NO \rightarrow (NO)_2$

$k_2 \hspace{1cm} (NO)_2 + RSH \rightarrow RSNO + HNO$

$k_3 \hspace{1cm}2 RSNO \rightarrow RSSR + (NO)_2$

$k_4\hspace{1cm} 2 HNO \rightarrow N_2O$

$k_5 \hspace{1cm} (NO)_2 + HNO \rightarrow HNO_2 + N_2O$

$k_6 \hspace{1cm} (NO)_2 \rightarrow 2 NO$

```python
reactions = [
    # (coeff, r_stoich, net_stoich)
    ('k1', {'[NO]': 2}, {'[NO]': -2, '[N2O2]': 1}),
    ('k2', {'[N2O2]': 1, '[RSH]': 1}, {'[N2O2]': -1, '[RSH]': -1, '[RSNO]': 1, '[HNO]': 1}),
    ('k3', {'[RSNO]': 2}, {'[RSNO]': -2, '[RSSR]': 1, '[N2O2]': 1}),
    ('k4', {'[HNO]': 2}, {'[HNO]': -2, '[N2O]': 1}),
    ('k5', {'[N2O2]': 1, '[HNO]':1}, {'[N2O2]': -1, '[HNO]': -1, '[HNO2]': 1, '[N2O]': 1}),
    ('k6', {'[N2O2]': 1}, {'[NO]': 2, '[N2O2]': -1})
]
names = '[NO] [RSH] [N2O2] [RSNO] [HNO] [RSSR] [HNO2] [N2O]'.split()
```

```python
from operator import mul
from functools import reduce
import sympy as sym

def prod(seq):
    return reduce(mul, seq) if seq else 1

def mk_exprs_symbs(rxns, names):
    # create symbols for reactants
    symbs = sym.symbols(names, real=True, nonnegative=True)
    # map between reactant symbols and keys in r_stoich, net_stoich
    c = dict(zip(names, symbs))
    f = {n: 0 for n in names}
    k = []
    for coeff, r_stoich, net_stoich in rxns:
        k.append(sym.S(coeff))
        r = k[-1]*prod([c[rk]**p for rk, p in r_stoich.items()])
        for net_key, net_mult in net_stoich.items():
            f[net_key] += net_mult*r
    return [f[n] for n in names], symbs, tuple(k)
```

genero el sistema de ecuaciones:

```python
sym.init_printing()
ydot, y, k = mk_exprs_symbs(reactions, names)
for i, ec in enumerate(ydot):
    print('d{0}/dt = {1}'.format(names[i], ec))
```

Construyo la funci√≥n a integrar y el jacobiano

```python
t = sym.symbols('t')
f = sym.lambdify((y, t) + k, ydot, 'numpy')
J = sym.Matrix(ydot).jacobian(y)
J_cb = sym.lambdify((y, t) + k, J, 'numpy')
```

Intento de ajuste de mecanismo a resultados experimentales

```python
def replace_fix_and_free(func, fixed_list):
    to_replace = [(ki, fixed_list[ki]) for ki in fixed_list.keys()]
    for i in range(len(func)):
        func[i] = func[i].subs(to_replace)
    return func
```

```python
def build_multiple_experiments(experiments, remove_list, ini_point, fin_point):
    list_of_experiments = set(experiments.keys())
    usable_experiments = list_of_experiments - set(remove_list)
    usable_experiments = sorted(list(usable_experiments))
    
    x_data = []
    y_data = []
    n_points = []
    
    for exp in usable_experiments:
        x_data.extend(experiments[exp].HNO[ini_point:fin_point].index.total_seconds().values)
        y_data.extend(experiments[exp].HNO[ini_point:fin_point].values)
        n_points.append(len(experiments[exp].HNO[ini_point:fin_point].index.total_seconds().values))
    
    x_data = np.array(x_data)
    y_data = np.array(y_data)
    return x_data, y_data, n_points, usable_experiments
```

Creo una funci√≥n que calcula la diferencia de cada experimento contra el modelo

```python
def leastsq_func(params, *args):
    n_points_per_experiment = args[0]
    x_data = args[1]
    y_data = args[2]
    conc_per_experiment = args[3]
    
    k_vals = [10**i for i in params]
    
    yfit = np.empty(y_data.shape)
    
    left = 0
    for i, val in enumerate(n_points_per_experiment):
        right = left + val
        yfit[left:right] = hno_t(x_data[left:right], k_vals, conc_per_experiment[i])
        left = right
    return y_data - yfit
```
